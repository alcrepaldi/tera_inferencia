```python
df = pd.read_csv('result_filtrado.csv')
```

```python
def grp_sint(df):
  if df['phq9'] < 5:
    return 0
  elif 5 <= df['phq9'] < 10:
    return 1
  elif 10 <= df['phq9'] < 15:
    return 2 
  elif 15 <= df['phq9'] < 19:
    return 3
  else:
    return 4
```

```python
df['phq_grp']=df.apply(grp_sint, axis=1)
df.sample(10)
```
